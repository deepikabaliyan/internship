# 2>Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
import requests
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import re
```

```python
url_2=requests.get('https://www.imdb.com/chart/top/')
url_2
```

```python
soup=BeautifulSoup(url_2.content)
```

```python
movie_name = []

for i in soup.find_all('td',class_="titleColumn"):
    movie_name.append(i.get_text().replace('\n', ""))

movie_name
```

```python
year = []

for i in soup.find_all('span',class_="secondaryInfo"):
    year.append(i.text)
year
        
 
```

```python
rating = []

for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text)
rating
```

```python
import pandas as pd
df = pd.DataFrame({"movie_name":movie_name,"year":year,"rating":rating})
df
```

# Q7. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world

i).   Headline
ii).  Time
iii)  News Link

```python
news=requests.get("https://www.cnbc.com/world/?region=world")
news
```

```python
soup=BeautifulSoup(news.content)
```

```python
headline= []
for i in soup.find_all('a',class_="LatestNews-headline"):
    headline.append(i.text)
headline
```

```python
time= []
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
time
```

```python
newsLink = []
for i in soup.find_all('a',class_="LatestNews-headline"):
    newsLink.append(i.get("href"))   
newsLink
```
